```python
# create a dataset with day_count  cloumns year , month and sorted Day with out duplicate

# Assuming the provided code has been executed and 'final_df' is available

day_count = final_df.groupby(['year', 'Month'])['Day'].nunique().reset_index()
day_count = day_count.rename(columns={'Day': 'total_days'})
day_count = day_count.sort_values(by=['year', 'Month', 'total_days']).reset_index(drop=True)

day_count
```
